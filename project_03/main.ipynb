```python
from patient_record_mgmt import PatientRecordManagementSystem
from graphviz import Digraph
```

```python
def main():
    patient_records = PatientRecordManagementSystem()
    patient_records.build_tree_from_csv('data\patient_records.csv')
    patient_2 = patient_records.search_patient_record(2)
    print(patient_2)
    patient_25 = patient_records.search_patient_record(25)
    print(patient_25.value)
    patient_47 = patient_records.search_patient_record(47)
    print(patient_47)

    dot_initial = patient_records.visualize_tree()
    dot_initial.render('patient_records_tree_initial.dot', view = False)
    
    patient_records.delete_patient_record(3)
    patient_records.delete_patient_record(10)
    patient_records.delete_patient_record(30)
    dot_after_deletion = patient_records.visualize_tree()
    dot_after_deletion.render('patient_records_tree_after-deletion.dot', view = False)

    new_records = [(51, "Jacob Marley", 40, "Migraine", "120/80", 70, 37.0),
                   (58, "Robert Shea", 55, "Back Pain", "130/85", 75, 36.8),
                   (3, "Joan Smith", 33, "Anxiety", "115/75", 68, 37.1)]
    for record in new_records:
        patient_records.add_patient_record(record)
    dot_after_insertion = patient_records.visualize_tree()
    dot_after_insertion.render('patient_records_tree_after-insertion.dot', view = False)

if __name__ == "__main__":
    main()
```
